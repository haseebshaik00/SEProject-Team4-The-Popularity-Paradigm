```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd

url = "https://incubator.apache.org/clutch/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
rows = soup.find_all('tr')

data_list = []
for row in rows[1:]:
    cols = row.find_all('td')
    data = {
        "Podling": cols[0].get_text(strip=True),
        "Sponsor": cols[1].get_text(strip=True),
        "start_date": cols[2].get_text(strip=True),
        "elapsed_days": cols[3].get_text(strip=True),
        "reporting_monthly": cols[4].get_text(strip=True),
        "reporting_group": cols[5].get_text(strip=True),
        "has_status_file": cols[6].get_text(strip=True),
        "status_update_date": cols[7].get_text(strip=True),
        "status_age": cols[8].get_text(strip=True),
        "status_update_counts": cols[9].get_text(strip=True),
        "no_of_committers": cols[10].get_text(strip=True),
        "no_of_new_committers": cols[11].get_text(strip=True),
        "has_a_code_repo": cols[12].get_text(strip=True),
        "has_issue_tracker": cols[13].get_text(strip=True),
        "has_a_dev_mailing_list": cols[14].get_text(strip=True),
        "has_a_commits_mailing_list": cols[15].get_text(strip=True),
        "has_website": cols[16].get_text(strip=True),
        "has_distribution_area": cols[17].get_text(strip=True),
        "has_release_signing_keys": cols[18].get_text(strip=True),
        "has_release_distribution_area": cols[19].get_text(strip=True),
    }
    data_list.append(data)

df = pd.DataFrame(data_list)
df = df.iloc[:-1]  # Remove the last row
display(df.head())
```

```python
import requests
import lxml.html as lx
import time
import json

podling_names = df['Podling'].tolist()
podlings = podling_names
BASE_URL = "https://incubator.apache.org/projects/"
HEADERS = {"User-Agent": "Mozilla/5.0"}

github_urls = {}
def get_github_url(podling):
    podling_url = f"{BASE_URL}{podling.lower()}.html"
    response = requests.get(podling_url, headers=HEADERS)

    if response.status_code == 200:
        tree = lx.fromstring(response.content)
        github_links = tree.xpath('//a[contains(@href, "github.com/apache")]/@href')

        if github_links:
            github_url = github_links[0]
            github_url = github_url.split("/issues")[0]
            return github_url
        else:
            return "GitHub URL not found"
    else:
        return "Page not found"

for podling in podlings:
    github_urls[podling] = get_github_url(podling)
    print(f"{podling}: {github_urls[podling]}")
    time.sleep(1)
```

```python
import requests
import json
import time
from config import GITHUB_TOKEN

HEADERS = {"Accept": "application/vnd.github.v3+json"}

def get_repo_stats(repo_url, github_token=None):
    if "github.com" not in repo_url:
        return {"stars": None, "forks": None, "pull_requests": None, "error": "Invalid URL"}
    parts = repo_url.replace("https://github.com/", "").strip("/").split("/")
    if len(parts) < 2:
        return {"stars": None, "forks": None, "pull_requests": None, "error": "Invalid repo format"}
    owner, repo = parts[:2]
    repo_api_url = f"https://api.github.com/repos/{owner}/{repo}"
    pr_api_url = f"https://api.github.com/repos/{owner}/{repo}/pulls?state=all"
    headers = HEADERS.copy()
    if github_token:
        headers["Authorization"] = f"Bearer {GITHUB_TOKEN}"
    repo_response = requests.get(repo_api_url, headers=headers)
    pr_response = requests.get(pr_api_url, headers=headers)
    if repo_response.status_code == 200 and pr_response.status_code == 200:
        repo_data = repo_response.json()
        pr_data = pr_response.json()
        pr_count = len(pr_data)
        stars = repo_data.get("stargazers_count", 0)
        forks = repo_data.get("forks_count", 0)
        pScore = stars + forks + (pr_count ** 2)
        return {"stars": stars, "forks": forks, "pull_requests": pr_count, "pScore": pScore}
    elif repo_response.status_code == 403 or pr_response.status_code == 403:
        return {"error": "Rate limit exceeded! Try again later or use a valid token."}
    else:
        return {"stars": None, "forks": None, "pull_requests": None, 
                "error": f"Failed to fetch data (Status {repo_response.status_code} / {pr_response.status_code})"}

def fetch_podling_data(podlings, github_token=None):
    podling_data = {}
    for podling, repo_url in podlings.items():
        podling_data[podling] = get_repo_stats(repo_url, github_token)
        print(f"Fetched: {podling} -> {podling_data[podling]}")
        time.sleep(1)
    with open("./datasets/7-podling_github_stats.json", "w") as f:
        json.dump(podling_data, f, indent=4)
    return podling_data

scraped_data = fetch_podling_data(github_urls, github_token=GITHUB_TOKEN)
```

```python
df_target = pd.DataFrame(scraped_data)
display(df_target)
```

```python
import pandas as pd

df_target_transposed = df_target.T.reset_index()
df_target_transposed.rename(columns={'index': 'Podling'}, inplace=True)
df_merged = df.merge(df_target_transposed, on='Podling', how='inner')
min_score = df_merged['pScore'].min()
max_score = df_merged['pScore'].max()
df_merged['pScore_normalized'] = (df_merged['pScore'] - min_score) / (max_score - min_score)
threshold = df_merged['pScore_normalized'].quantile(0.3)
df_merged['Popular'] = df_merged['pScore_normalized'].apply(lambda x: 1 if x >= threshold else 0)
df_merged_sorted = df_merged.sort_values(by='pScore_normalized', ascending=False).reset_index(drop=True)

display(df_merged_sorted.head())
df_merged_sorted.to_csv('./datasets/final-rq3-dataset.csv', index=False)
print("Data stored in 'final-rq3-dataset.csv'!")
```
